# Part I: Making a Basic Data Call

```python
# Install and import packages 
import quandl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Retrieve unfiltered time-series
quandl.ApiConfig.api_key = 'YOURAPIKEY'
Apple_EOD = quandl.get('EOD/AAPL')
Apple_EOD.head()
```

```python
# Plot the adjjusted close price over time using pandas plot function
Apple_EOD.plot(y= 'Adj_Close', figsize=(12, 11))
plt.xlabel('Date', fontsize=18)
plt.ylabel('Adj_Close', fontsize=16)
plt.title("Apple End Of Day Price")
plt.savefig('test.jpg')
plt.show()
```

```python
# Retrieve fundamentals ratios table for Apple
Apple_FR = quandl.get_table('ZACKS/FR', ticker='AAPL')
Apple_FR.head()
```

# Part II: Manipulating Time-Series Data

```python
Apple_EOD_2= quandl.get('EOD/AAPL', start_date = "2015-12-12", end_date = "2019-12-30")
Apple_EOD_2.head()
```

```python
Apple_EOD_3 = quandl.get('EOD/AAPL', start_date = "2015-12-12", end_date = "2019-12-30", collapse="monthly")
Apple_EOD_3.head()
```

```python
Apple_EOD_col11 = quandl.get('EOD/AAPL', start_date = "2015-12-12", end_date = "2019-12-30", column_index=11)
Apple_EOD_col11.head()
```

```python
Apple_pct_change = quandl.get('EOD/AAPL',start_date ="2015-12-12",end_date = "2019-12-30",column_index=11,
                              collapse="monthly",transformation="diff")
Apple_pct_change.head(10)
```
